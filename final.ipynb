```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from sklearn.preprocessing import LabelEncoder
from matplotlib.colors import ListedColormap
from matplotlib import colors
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("marketing_campaign.csv", sep="\t")
df
```

```python
df['Education'].unique()
```

```python
import matplotlib.pyplot as plt

# Extract the categorical column
categorical_data = df['Education']

# Calculate the frequency of each category
category_counts = categorical_data.value_counts()

# Plot the histogram (bar chart for categorical data)
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title('Frequency Distribution of Marital Status')
plt.xlabel('Education')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
